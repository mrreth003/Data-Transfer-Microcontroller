```python
from matplotlib import pyplot as plt
import numpy as np
from numpy.fft import fft, ifft
import csv
import math

timeArr =[] #array of all the times in the file
accel = []
tempArr = []
gyrX = []
gyrY = []
gyrZ = []

with open("./delay100.csv", 'r') as file:
    csvreader = file.readlines()
    i = 0
    for line in csvreader:
        if i==0:
            i = 1
        else:
            splitRow = line.split(",")
            tempArr.append(float(splitRow[0]))    
            
            gyrX.append(float(splitRow[1]))
            gyrY.append(float(splitRow[2]))
            gyrZ.append(float(splitRow[3]))
            
            accelX = float(splitRow[4])
            accelY = float(splitRow[5])
            accelZ = float(splitRow[6])
            accelMag = math.sqrt((accelX**2)+(accelY**2)+(accelZ**2))
            accel.append(accelMag)
                        
    # fft of the time v temperature plot:
    sr = 10000 # Sampling rate
    X = fft(tempArr)
    N = len(X)
    n = np.arange(N)
    T = N/sr
    freq = n/T 
    plt.figure(figsize=(8, 2))
    plt.plot(freq, np.abs(X))
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('FFT Amplitude |Temperature(f)|')
    plt.show()

    # fft of the time v accleration plot:
    Y = fft(accel)
    Z = len(Y)
    n1 = np.arange(Z)
    Tt = Z/sr
    freq = n1/Tt 
    plt.figure(figsize=(8, 2))
    plt.plot(freq, np.abs(Y))
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('FFT Amplitude |Acceleration(f)|')
    plt.show()
    
    # fft of the time v gryX plot:
    Y = fft(gyrX)
    Z = len(Y)
    n1 = np.arange(Z)
    Tt = Z/sr
    freq = n1/Tt 
    plt.figure(figsize=(8, 2))
    plt.plot(freq, np.abs(Y))
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('FFT Amplitude |Rotation(f)|')
    plt.show()
    
file.close()
```
